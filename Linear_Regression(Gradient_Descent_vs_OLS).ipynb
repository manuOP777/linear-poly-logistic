```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```
epochs = 5000

#initialising m and c
m=0
c =0
learning_rate=0.0001
n= float(len(X))

prev_error=100

for i in range(epochs):
  y_pred = m *X + c
  m_stepsize = learning_rate * ((-2/n) *sum(X * (y-y_pred)))
  c_stepsize = learning_rate * ((-2/n) *sum(y-y_pred))
  m= m - m_stepsize
  c =c + c_stepsize
  error = (1/n) * sum((y-y_pred)**2)
  if error>=prev_error:
    print("epoch count=" , i)
    print("error =",error)
    break
  else:
    prev_error =error
  

  

print("m value =",m)
print("c value = " ,c)
```

```
y_pred = m*X +c 

plt.scatter(X,y)
plt.plot(X,y_pred)
```

USING LEAST SQUARE METHOD

But this solution is not scalable. Applying this to Linear Regression was fairly easy as we had nice coefficients and linear equations. Applying this to complex and non-linear algorithms like Support Vector Machine will not be feasible. So we will find the numerical approximation of this solution by iterative method  which would be close to (but not exactly equal to) the OLS solution  which gave us the exact solution.


```
X_mean= np.mean(X)
y_mean = np.mean(y)


m = sum((X-X_mean)*(y-y_mean))/sum((X-X_mean)**2)


c = y_mean - m *X_mean


print("m value =",m)
print("c value = " ,c)

y_pred = m*X+c

error = (1/n) * sum((y-y_pred)**2)

print("error=",error)
```
